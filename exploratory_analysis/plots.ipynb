```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# load the dataset
df = pd.read_csv('../code/temp_data_v2/before_model.csv')
```

```python
# plot the distribution of the target variable TRIAJ_New, which is the new triage level
# Red: Emergent, Yellow: Urgent, Yeşil: Less Urgent
# name of y axis is changed to 'Number of Patients' for better visualization
# name of x axis is changed to 'Triage Level' for better visualization
plt.figure(figsize=(10, 6))
sns.countplot(x='TRIAJ_New', data=df)
plt.ylabel('Number of Patients')
plt.xlabel('Triage Level')
plt.title('Distribution of Triage Level')
plt.show()
```

```python
cols = ['YASI', 'Nabız',
       'Tansiyon_Sistolik', 'Tansiyon_Diastolik', 'Oksijen_Satürasyonu',
       'Ateş']
cols_tr = ['Age', 'Pulse', 'Systolic', 'Diastolic', 'Oxygen', 'Temperature']
count = 0
for col in cols:
    count += 1
    plt.figure(figsize=(10, 6))
    plt.subplot(2, 1, 1)
    plt.hist(df_first[(df_first[col]>df_first[col].quantile(0.01))&(df_first[col]<df_first[col].quantile(0.99))][col], bins=40)
    plt.axvline(df_first[col].quantile(0.25), color='red', linestyle='dashed', linewidth=2, label='25%')
    plt.axvline(df_first[col].quantile(0.5), color='black', linestyle='dashed', linewidth=2, label='50%')
    plt.axvline(df_first[col].quantile(0.75), color='red', linestyle='dashed', linewidth=2, label='75%')
    plt.ylabel('Number of Patients')
    plt.title('Distribution of '+cols_tr[count-1]+' Before Preprocessing')
    plt.xlim(df_first[col].quantile(0.01), df_first[col].quantile(0.99))
    plt.legend()
    # add value of percentiles to the plot as text in mid of vertical lines
    plt.text(df_first[col].quantile(0.25), 100, df_first[col].quantile(0.25))
    plt.text(df_first[col].quantile(0.5), 100, df_first[col].quantile(0.5))
    plt.text(df_first[col].quantile(0.75), 100, df_first[col].quantile(0.75))
    plt.subplot(2, 1, 2)
    plt.hist(df[(df[col]>df[col].quantile(0.01))&(df[col]<df[col].quantile(0.99))][col], bins=40)
    plt.axvline(df[col].quantile(0.25), color='red', linestyle='dashed', linewidth=2, label='25%')
    plt.axvline(df[col].quantile(0.5), color='black', linestyle='dashed', linewidth=2, label='50%')
    plt.axvline(df[col].quantile(0.75), color='red', linestyle='dashed', linewidth=2, label='75%')
    plt.ylabel('Number of Patients')
    plt.xlabel(cols_tr[count-1])
    plt.title('Distribution of '+cols_tr[count-1]+' after Preprocessing')
    plt.xlim(df[col].quantile(0.01), df[col].quantile(0.99))
    plt.legend()
    # add value of percentiles to the plot as text in mid of vertical lines
    plt.text(df[col].quantile(0.25), 100, df[col].quantile(0.25))
    plt.text(df[col].quantile(0.5), 100, df[col].quantile(0.5))
    plt.text(df[col].quantile(0.75), 100, df[col].quantile(0.75))
    plt.show()
    
    
```

```python
# show the distribution of Ateş in df for each triage level as a table
# add %25, %50, %75 values
# name of y axis is changed to 'Number of Patients' for better visualization
# name of x axis is changed to 'Temperature' for better visualization
# Add value of percentiles to the plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='TRIAJ_New', y='Ateş', data=df[(df['Ateş']>35)&(df['Ateş']<42)], whis=[25, 75])
plt.ylabel('Temperature')
plt.xlabel('Triage Level')
plt.title('Distribution of Temperature for Each Triage Level')
plt.show()
```

```python
# create a table that shows the statistical summary of df and df_first
# mean, median, %25, %50, %75 values of Ateş for each triage level is added to the table
df_first.Ateş.describe()
df.Ateş.describe()
```

```python
# create a table that shows the statistical summary of df and df_first for each triage level
for col in cols:
    mean_before = df_first[col].describe()[1]
    mean_after = df[col].describe()[1]
    median_before = df_first[col].describe()[5]
    median_after = df[col].describe()[5]
    q1_before = df_first[col].describe()[4]
    q1_after = df[col].describe()[4]
    q3_before = df_first[col].describe()[6]
    q3_after = df[col].describe()[6]
    # create a pd dataframe to show the statistical summary of df and df_first
    print(pd.DataFrame({'Before Preprocessing': [mean_before, median_before, q1_before, q3_before],col: ['Mean', 'Median', '25%', '75%'], 'After Preprocessing': [mean_after, median_after, q1_after, q3_after]}).set_index(col).T)
```
